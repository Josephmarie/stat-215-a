```python
import pandas as pd
import plotly.express as px
```

```python
# read data from csv
clean_ling_data = pd.read_csv('../data/ling_data.csv')
clean_lingloc = pd.read_csv('../data/clean_lingloc.csv')
```

***2 questions I choose***

***Question 50***:What word(s) do you use to address a group of two or more people?
This question can reveal how people in different regions address a group of people, which can vary from region to region and can be related to the local culture or dialect.

***Question 53***:Modals are words like "can," "could," "might," "ought to," and so on. Can you use more than one modal at a time? (e.g.,"I might could do that" to mean "I might be able to do that"; or "I used to could do that" to mean "I used to be able to do that")
This question explores a particular construct in language use that has different levels of popularity in different English dialects. It can help reveal the geographical distribution of language habits.

```python
eda_df = clean_ling_data[['ID', 'CITY', 'STATE', 'ZIP', 'Q050', 'Q053', 'lat', 'long']]
```

```python
fig = px.scatter_mapbox(eda_df, lat='lat', lon='long', color='Q050',width=850, height=700,
                            size='Q053',
                            size_max=5,
                            hover_data=['CITY', 'STATE', 'ZIP', 'Q050'],
                            mapbox_style="open-street-map",
                            title="Survey Responses (Q50 and Q53) on a Map in North America",
                            center={"lat": 45, "lon": -100}, # Set the map center to the general area of North America
                            zoom=3)  # Set the initial zoom level

fig.show()
```
